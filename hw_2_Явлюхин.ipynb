## Примеры:

- "Петров;Сергей;1970-11-05;A15.0;активный" - # True
- "Сидорова;Мария1;1990-01-30;B20;ремиссия" - # False (в имени цифра)
- "Кузнецов;Дмитрий;1988-13-01;A16;выписан" - # False (некорректный месяц)
- "Васильева;Елена;2000-02-29;C01.1;наблюдение" - # False (неверный статус)

```python
patient1 = "Петров;Сергей;1970-11-05;A15.1;активный"
patient2 = 'Сидорова;Мария1;1990-01-30;B20;ремиссия'
patient3 = 'Кузнецов;Дмитрий;1988-13-01;A16;выписан'
patient4 = 'Васильева;Елена;2000-02-29;C01.1;наблюдение'

def check_correctness(entry):
    parts = entry.split(';')
    if len(parts) != 5: return False

    surname = parts[0]
    name = parts[1]

    dob = parts[2].split('-')
    if len(dob) != 3: return False
    for i in dob:
        if i.isdecimal() == False: return False

    year = dob[0]
    month = dob[1]
    day = dob[2]

    diagnosis = parts[3]
    split_diagnosis = diagnosis.split('.')
    if not (1 <= len(split_diagnosis) <=2): return False
    if len(split_diagnosis) == 2:
        if len(split_diagnosis[1]) != 1: return False
    status = parts[4]

    return name.isalpha() and name.istitle() and not (' ' in name)\
        and surname.isalpha() and surname.istitle() and not (' ' in surname)\
        and len(year) == 4 and len(month) == 2 and len(day) == 2\
        and (1 <= int(month) <= 12) and (1 <= int(day) <= 31)\
        and diagnosis.startswith('A' or 'B' or 'C') \
        and 2 <= len(split_diagnosis[0][1:]) <= 3 \
        and (status == 'активный' or status == 'ремиссия' or status == 'выписан')

print(check_correctness(patient1))
print(check_correctness(patient2))
print(check_correctness(patient3))
print(check_correctness(patient4))
```
