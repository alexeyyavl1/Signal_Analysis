```python
patient_weight_1 = 82.5
patient_height_1 = 186
patient_age_1 = 34
patient_weight_2 = 95
patient_height_2 = 190
patient_age_2 = 42
```

```python
print('Вес второго пациента превышает первого в', patient_weight_2 / patient_weight_1, 'раз.')
print('Рост второго пациента превышает первого в', patient_height_2 / patient_height_1, 'раз.')
```

```python
BMI_1 = patient_weight_1 / (patient_height_1 / 100) ** 2
BMI_2 = patient_weight_2 / (patient_height_2 / 100) ** 2
print('ИМТ пациента 1:', BMI_1)
print('ИМТ пациента 2:', BMI_2)
```

```python
bpm_1 = 70 * 60 * 24 * patient_age_1
bpm_2 = 70 * 60 * 24 * patient_age_2
print('Приблизительное количество ударов сердца за период жизни у пациента 1:', bpm_1)
print('Приблизительное количество ударов сердца за период жизни у пациента 2:', bpm_2)
```

Две ЭКГ-шки:
ID123_CH05_F32.5_A0.001_B0.002
ID456_CH12_F40.0_A0.003_B0.001

```python
file1 = "ID123_CH05_F32.5_A0.001_B0.002"
file2 = "ID456_CH12_F40.0_A0.003_B0.001"
```

```python
def make_readable(ecg):
    print(ecg)
    parts = ecg.split('_')
    id = int(parts[0][2:])
    chan = int(parts[1][2:])
    freq = float(parts[2][1::])
    amp = float(parts[3][1::])
    noise = float(parts[4][1::])
    print(f"ID: {id} \nChannel: {chan} \nFrequency: {freq} \nAmplitude: {amp} \nNoise: {noise}\n------------------------------")
```

```python
make_readable(file1)
make_readable(file2)
```
