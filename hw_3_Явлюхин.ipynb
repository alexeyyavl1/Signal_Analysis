```python
temperatures = [36.6, 36.7, 36.8, 36.9, 37.1, 37.3, 37.5, 37.4, 37.2, 37.0, 36.9, 36.8]
times = []

for i in range(12):
    if i * 2 < 10:
        times.append('0' + str(i*2) + ':00')
    else:
        times.append(str(i*2) + ':00')

def evaluate_temperatures(temp):
    max_temperature = max(temp)
    if max_temperature <= 37.1: return "Температурный режим в пределах нормы"
    print(f'Максимальная температура: {max_temperature} в {times[temperatures.index(max_temperature)]}')

    subfebrile = []
    for i in range(12):
        if temperatures[i] >= 37.1:
            subfebrile.append([temperatures[i], times[i]])

    print(f'Количество измерений выше 37.0°C: {len(subfebrile)}')
    print('Подозрительные измерения:')
    for i in range(len(subfebrile)):
        print(f' - {subfebrile[i][0]}°C в {subfebrile[i][1]}')

    if len(subfebrile) >= 3:
        print('ПОДОЗРЕНИЕ НА АКТИВНЫЙ ТУБЕРКУЛЁЗ')
    else:
        print('Температурный режим в пределах нормы')

evaluate_temperatures(temperatures)
```
